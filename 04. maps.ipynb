### Libraries

```python
import folium
from folium import Choropleth, Circle, Marker, Icon, Map
from folium.plugins import HeatMap, MarkerCluster
import pandas as pd
```

```python
df = pd.read_csv('dataset/kytecompany.csv')
df
```

📍

```python
lat = 37.788482
lon = -122.409173
```

```python
map_ = Map(location=[37.788482,-122.409173],zoom_start=15)
map_
```

```python
icono = Icon(color = "blue",
             prefix = "fa",
             icon = "briefcase",
             icon_color = "black")

loc = {"location":[lat, lon],
      "tooltip": "Casa de Dobby"}
```

```python
marker_cueva = Marker(**loc, icon = icono)
```

```python
marker_cueva.add_to(map_)
map_
```

```python
for i, row in df.iterrows():
    distrito = {
        "location":[row["latitude"], row["longitude"]]
    }
    
    if row["local"] == "starbucks":
        icon = Icon(color = "green",
                    prefix = "fa",
                    icon = "coffee",
                    icon_color = "black"
        )
        
    elif row["local"] == "vegan":
        icon = Icon(color = "blue",
                    prefix = "fa",
                    icon = "cutlery",
                    icon_color = "black"
        )
        
    else:
        icon = Icon(color = "red",
                    prefix = "fa",
                    icon = "plane",
                    icon_color = "black"
        )
    Marker(**distrito,icon = icon ).add_to(map_)
map_
```
